```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.gaussian_process.kernels import RBF
from scipy.linalg import cholesky

# 1. Create 1D input grids
x1 = np.linspace(-1, 1, 25).reshape(-1, 1)
x2 = np.linspace(-1, 1, 25).reshape(-1, 1)


X1_grid, X2_grid = np.meshgrid(x1, x2)  # for plotting only

# 2. Define 1D RBF kernels
length_scale = 0.3
rbf = RBF(length_scale=length_scale)
K1 = rbf(x1, x1)  # (50x50)
K2 = rbf(x2, x2)  # (50x50)

# 3. Compute outer products (surface plots of kernels)
K1_plot = np.outer(K1[:, 12], np.ones(25))  # Fix x1' = x1[25]
K2_plot = np.outer(np.ones(25), K2[:, 12])  # Fix x2' = x2[25]
K_product_plot = K1_plot * K2_plot

# 4. Build full product kernel via Kronecker product
K_product = np.kron(K1, K2)

# 5. Sample from GP
mean = np.zeros(K_product.shape[0])
cov = K_product + 1e-6 * np.eye(K_product.shape[0])  # add jitter for stability
gp_sample = np.random.multivariate_normal(mean, cov).reshape(25, 25)

# 6. Plotting
fig = plt.figure(figsize=(16, 4))

def plot_surface(ax, X, Y, Z, title):
    ax.plot_surface(X, Y, Z, cmap="viridis", edgecolor='none')
    ax.set_title(title, fontsize=12)
    ax.set_xlabel('$x_1$')
    ax.set_ylabel('$x_2$')

# Subplots
ax1 = fig.add_subplot(1, 4, 1, projection='3d')
plot_surface(ax1, X1_grid, X2_grid, K1_plot, r"$\mathrm{SE}_1(x_1, x_1')$")

ax2 = fig.add_subplot(1, 4, 2, projection='3d')
plot_surface(ax2, X1_grid, X2_grid, K2_plot, r"$\mathrm{SE}_2(x_2, x_2')$")

ax3 = fig.add_subplot(1, 4, 3, projection='3d')
plot_surface(ax3, X1_grid, X2_grid, K_product_plot, r"$\mathrm{SE}_1 \times \mathrm{SE}_2$")

ax4 = fig.add_subplot(1, 4, 4, projection='3d')
plot_surface(ax4, X1_grid, X2_grid, gp_sample, r"Sample from $\mathcal{GP}(0, \mathrm{SE}_1 \times \mathrm{SE}_2)$")

plt.tight_layout()
#plt.savefig(r"C:\Users\seanw\OneDrive\Desktop\GitHub\FYP\Latex\LatexPlots\final_gps_plots\2dkernels",dpi=72, bbox_inches='tight')
plt.show()
```
